```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.options.display.max_rows = None
pd.options.display.max_columns = None
pd.set_option("display.precision", 8)

n_rows=5e2

Liczby=pd.DataFrame(np.random.normal(size=(int(n_rows),3)),columns=["A","B","C"])
display(Liczby)
Liczby.describe()
```

```python
kolory_hex = ['#000080']

plt.figure()

plt.scatter(Liczby['A'], Liczby['B'], c=kolory_hex)
plt.xlabel('A')  
plt.ylabel('B')

plt.xlim(-6, 6)
plt.ylim(-4, 4)

plt.title('Wykres punktowy kolumn A i B') 
plt.colorbar(label='B')  
plt.show()  
```

```python
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

kolory_hex = ['#000080']

mappable = ax.scatter(Liczby['A'], Liczby['B'], Liczby['C'], c=kolory_hex)

ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('C')

ax.set_xlim(-4, 4)  
ax.set_ylim(-4, 4)  
ax.set_zlim(-4, 4)  

ax.set_title('Wykres punktowy 3D kolumn A, B i C')

plt.show()
```
